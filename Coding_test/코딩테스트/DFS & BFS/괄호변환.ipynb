```python
def str_to_list(p):
    p_list = []
    if len(p) == 0:
        return p_list
    
    for s in p:
        p_list.append(s)
    
    return p_list

def uvsplit(p):
    right = 0
    left = 0
    u = []
    if len(p) == 0:
        return u, u
    for s in p:
        if s == '(':
            left +=1
            u += s
        if s == ')':
            right += 1
            u += s
        if right == left:
            return u, p[len(u):]

def check(p):
    right = 0
    left = 0
    for s in p:
        if s == '(':
            left +=1
        if s == ')':
            right += 1
        if left < right:
            return False
    
    return True

def solution(p):
    answer = ''
    
    if p == '':
        return answer
    
    u, v = uvsplit(p)
    u = str_to_list(u)
    v = str_to_list(v)
    
    if check(u):
        answer = u + solution(v)
    
    else:
        answer = ['(']
        answer += solution(v)
        answer += [')']
        
        for i in range(len(u)):
            if u[i] == '(':
                u[i] = ')'
            else:
                u[i] = '('
        
        u = u[1:-1]
        
        answer = u + answer
    
    str_answer = ""
    
    for s in answer:
        str_answer+=s
        
    
    return str_answer
```

```python
def balanced_index(p):
    count = 0
    
    for i in range(len(p)):
        if p[i] == '(':
            count += 1
        
        else:
            count -= 1
        
        if count == 0:
            return i

def check_proper(p):
    count = 0
    
    for i in p:
        if i == '(':
            count +=1
        
        else:
            if count == 0:
                return False
            count -= 1
    
    return True

def solution(p):
    answer = ''
    if p == '':
        return answer
    
    index = balanced_index(p)
    
    u = p[:index+1]
    v = p[index+1:]
    
    if check_proper(u):
        answer = u + solution(v)
    
    else:
        answer = '('
        answer += solution(v)
        answer += ')'
        
        u = list(u[1:-1])
        
        for i in range(len(u)):
            if u[i] == '(':
                u[i] = ')'
            
            else:
                u[i] = '('
            
        answer += "".join(u)
    
    return answer
```
